```python
!pip install ultralytics --quiet
print("YOLOv8 Installation Complete.")
from google.colab import drive
drive.mount('/content/drive')
print("Google Drive Mounted.")
import zipfile, os
zip_path = "/content/drive/MyDrive/archive (2).zip"`
extract_path = "/content/yolo_data"
os.makedirs(extract_path, exist_ok=True)

# The above are the set of code to mount google drive and apply the path of my dataset ( You can find the dataset in kaggle although I provided it below)
# The dataset used is - https://www.kaggle.com/c/byu-locating-bacterial-flagellar-motors-2025
# You can even directly use kaggle hub and download the dataset instantly everytime u train !


print(f"Unzipping dataset from {zip_path}...")
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print(f" Dataset unzipped to {extract_path}")


#   Create the dataset.yaml file
# This file tells YOLOv8 where to find the training/validation images and what the class names are.
yaml_path = os.path.join(extract_path, 'dataset.yaml')

with open(yaml_path, 'w') as f:
    f.write(f"""
train: {extract_path}/yolo_dataset/images/train
val: {extract_path}/yolo_dataset/images/val

# Number of classes
nc: 2
names: ['motor0', 'motor1']
""")
print(f" dataset.yaml created at {yaml_path}")
from ultralytics import YOLO
model = YOLO("yolov8l.pt")

# Define a more descriptive name for the training run.
run_name = "motor_detector_v8x_high_accuracy"

print("Starting model training.")
# more the epoch the better training but keep patience level more though for good result and in order to save time for future training.
# It will surely overlift In my case as I have kept the patience level at 10. ( A dry run )
results = model.train(
    data=yaml_path,
    epochs=200,          # no of epoch to be trained
    imgsz=940,          # sizeee of img
    batch=6,           # batch size
    patience=10,        # patience level
    name=run_name,      # output folder name
    device=0,           # Use GPU.( Single gpu used so 0)
    task='detect',      # task i performed
    #  Data Augmentation
    degrees=10,         # Image rotation
    translate=0.1,      # Image translation
    scale=0.2,          # Image scaling
    fliplr=0.5,         # horizontal flip
    mosaic=1.0          # Use mosaic augmentation
)
print(" Training Complete.")


# Load the best performing model saved during training.
best_model_path = f"runs/detect/{run_name}/weights/best.pt"
model = YOLO(best_model_path)

# Run validation and print the metrics. This will show the final mAP50-95 score
print(" Validating the best model...")
metrics = model.val(data=yaml_path)
# These are my metrics . These are the standard metrics while training any yolo model .

print("\n--- Validation Metrics ---")
print(f"mAP50-95: {metrics.box.map:.4f}")
print(f"   mAP50: {metrics.box.map50:.4f}")
print(f"   mAP75: {metrics.box.map75:.4f}")
print("--------------------------\n")


#  Plot Training Metrics
metrics_path = f"runs/detect/{run_name}/results.png"
from IPython.display import Image
print(" Displaying training metrics graph:")
display(Image(filename=metrics_path))


#  Visualize Predictions on Random Validation Images
import random
import glob
import matplotlib.pyplot as plt

# Get a list of all validation images
val_images = glob.glob(f"{extract_path}/yolo_dataset/images/val/*.jpg")
# Select 3 random images to test
samples = random.sample(val_images, min(3, len(val_images))) # Ensuringg we don't exceed list size

print("\n Visualizing predictions on 3 random validation images...")
for img_path in samples:
    # Make predictions with a confidence threshold
    preds = model.predict(img_path, save=False, conf=0.5)
    # Plot the image with bounding boxes
    pred_img = preds[0].plot()

    # Display the image using matplotlib
    plt.figure(figsize=(12, 8))
    # BGR to RGB conversion for correct color display
    plt.imshow(pred_img[:, :, ::-1])
    plt.title(f"Predictions: {os.path.basename(img_path)}")
    plt.axis('off')
    plt.show()
```

The above are the one of the three predicted images which has the binding boxes.. the number with it is the confidence score the confidence score >=0.70 is considered the best score in yolo projects for the research grade and can be improved by augmentations.
